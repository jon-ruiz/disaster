```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
# %matplotlib.inline
```

```python
df = pd.read_pickle('../tweets_cleaned.pkl')
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.head()
```

```python
text = ' '.join(tweet for tweet in df.tweet_text)
```

```python
text = text.replace('USERID','')
```

```python
text = text.replace('HTTP','')
```

```python
text = text.replace('\\','')
```

```python
text = text.replace("&gt","")
```

```python
# NOTE: By it looks like this does NOT tokenize, so we will have to do 
       # that separately and run it on the tokenized text
# Generate a word cloud image
wordcloud = WordCloud(width=1000,height=500,scale=1,background_color="white").generate(text)

# Display the generated image:
# the matplotlib way:
# import matplotlib.pyplot as plt
# plt.imshow(wordcloud, interpolation='bilinear')
# plt.axis("off")

# lower max_font_size
wordcloud = WordCloud(width=1000,height=500,scale=1,background_color="white").generate(text)
plt.figure(figsize=(15,10))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
